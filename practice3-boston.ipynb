### 0. Tensorflow & Pandas ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÇ¨Ïö©

```
import tensorflow as tf
import pandas as pd
```

### 1-1. Í≥ºÍ±∞Ïùò Ï∂ïÏ†ÅÎêú Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ

```
Bt_fileroute = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
Boston = pd.read_csv(Bt_fileroute)
print(Boston.columns)
```

### 1-2. ÎèÖÎ¶ΩÎ≥ÄÏàò, Ï¢ÖÏÜçÎ≥ÄÏàò Î∂ÑÎ¶¨ 

```
inde_var = Boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
            'ptratio', 'b', 'lstat']]
de_var = Boston[['medv']]
print(inde_var.shape, de_var.shape)
```

### 2. Î™®Îç∏ Íµ¨Ï°∞ ÏÉùÏÑ±

```
X = tf.keras.layers.Input(shape=[13]) # 13: ÎèÖÎ¶ΩÎ≥ÄÏàò Í∞úÏàò
Y = tf.keras.layers.Dense(1)(X) # 1: Ï¢ÖÏÜçÎ≥ÄÏàò Í∞úÏàò
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

### 3-1.Îç∞Ïù¥ÌÑ∞Î°ú Î™®Îç∏ ÌïôÏäµ(FIT)
ÌïôÏäµÏù¥ Ïûò Îê†ÏàòÎ°ù 0Ïóê Í∞ÄÍπåÏõåÏßê | ÏõêÌïòÎäî lossÍ∞íÏúºÎ°ú Îñ®Ïñ¥Ïßà ÎïåÍπåÏßÄ epoch Îçî ÌïòÍ∏∞

```
model.fit(inde_var, de_var, epochs=10000, verbose=0)
```

### 3-2. Î™®Îç∏Ïùò lossÍ∞í ÌôïÏù∏

```
model.fit(inde_var, de_var, epochs=10)
```

### 4-1. Î™®Îç∏ Ïù¥Ïö©ÌïòÏó¨ ÏòàÏ∏°

```
print(model.predict(inde_var[5:10]))
```

### 4-2. Ïã§Ï†úÍ∞í & ÏòàÏ∏°Í∞í ÎπÑÍµê
epochs Í∞í Îçî ÎäòÎ¶¨Í∏∞ üòì
