### 0. Tensorflow & Pandas 라이브러리 사용

```
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

### 1-1. 과거의 축적된 데이터 준비

```
# 1.과거의 데이터를 준비합니다.
file_route = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(file_route)
iris.head()
```

### 1-2. onehot-encoding
독립변수(x1 ~ x4) 4개, 종속변수(y) 1개_인데 범주형 data. <br>
y = w1x1 + w2x2 + w3x3 + w4x4 + b 등의 수식으로 표현 불가. <br>
→ HOW? 수식으로 사용할 수 있도록 바꿔주는 과정이 필요
****
모든 범주 칼럼으로 나타낸 뒤 해당되는 값에 1씩 ++ → 0 or 1: 원핫인코딩

```
iris_encoding = pd.get_dummies(iris) #범주형만 골라서 원핫인코딩 시킴
iris_encoding.head()
```

#### 1-2-1.칼럼 이름 출력
by. 원핫인코딩 → 독립변수 4개 & 종속변수 3개

```
print(iris_encoding.columns)
```

###1-3. 독립변수, 종속변수 분리 

```
inde_var = iris_encoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
de_var = iris_encoding[['품종_setosa', '품종_versicolor', '품종_virginica']]
print(inde_var.shape, de_var.shape)
```

### 2. 모델 구조 생성
****
<b>- Softmax?</b> <br>
대부분의 분류 예측: 0~100의 % 수치로 표현 <br>
기존 회귀형 수식에 softmax 함수(이런 함수s: Activation, 활성화 함수)를 씌워서 0부터 1 사이의 값으로 예측하도록 함.<br>
회귀모델: Identity (y = x) <br>
분류모델: Softmax <br>
※ Sigmoid란 것도 있음_ 차이점 알아봐 <br>
<b>- Crossentropy?</b> <br>
문제에 따라 loss를 다르게 사용해야 함. 분류에 사용하는 loss. 회귀는 loss='mse'<br>
accuracy도 알 수 있음(1에 가까울수록 정확도 ↑)


```
X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation='softmax')(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

### 3-1.데이터로 모델 학습(FIT)

```
model.fit(inde_var, de_var, epochs=1000, verbose=0)
```

### 3-2. 모델의 loss값 확인

```
model.fit(inde_var, de_var, epochs=10)
```

### 4-1. 모델 이용하여 예측

```
# 맨 처음 데이터 5개
print(model.predict(inde_var[:5]))
print(de_var[:5])

# 중간 데이터 5개
print(model.predict(inde_var[48:53]))
print(de_var[48:53])

# 맨 마지막 데이터 5개
print(model.predict(inde_var[-5:]))
print(de_var[-5:])
```

### 4-2. 모델의 수식 확인(weights & bias 출력)

```
print(model.get_weights())
```
